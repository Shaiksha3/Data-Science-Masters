# Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to specific functions or views that generate the appropriate response. This is done using the @app.route() decorator, which is used to associate a URL pattern with a function that handles requests to that URL.
